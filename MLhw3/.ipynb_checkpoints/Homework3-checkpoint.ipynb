```python
from __future__ import division, print_function, unicode_literals
from sklearn import metrics
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from scipy.spatial import distance


dataTrain=np.loadtxt(fname='ridgetrain.txt',dtype=float,delimiter=None)
dataTest=np.loadtxt(fname='ridgetest.txt',dtype=float,delimiter=None)
#X=data[:,0]
#X=np.reshape(X,(250,1))
#Xt=np.reshape(X,(1,250))
#XXt=np.dot(X,Xt)
K=np.empty((dataTrain.shape[0],dataTrain.shape[0]))
for i in range(0,dataTrain.shape[0]):
    for j in range(0,dataTrain.shape[0]):
        K[i][j]=np.exp(-0.1*((np.abs(dataTrain[i][0]-dataTrain[j][0]))**2))
I=np.identity(dataTrain.shape[0])
for lambda_ in [0.1,1,10,100]:
    alpha=np.multiply(lambda_,I)
    alpha=np.add(alpha,K)
    alpha=np.linalg.inv(alpha)
    alpha=np.dot(alpha,dataTrain[:,1])
    predictedy=np.zeros(dataTest.shape[0])
    for i in range(0,dataTest.shape[0]):
        for j in range(0,dataTrain.shape[0]):
            k=np.exp(-0.1*((np.abs(dataTrain[j][0]-dataTest[i][0]))**2))
            predictedy[i]=predictedy[i]+np.multiply(alpha[j],k)
    fig = plt.figure(figsize=(8,8))
    actual=matplotlib.pyplot.scatter(dataTest[:,0], dataTest[:,1], c='blue')
    predicted=matplotlib.pyplot.scatter(dataTest[:,0], predictedy, c='red')
    plt.legend((actual,predicted),('True output','Predicted output'),loc='upper left')
    plt.title('Plot of training and test points')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    rmse=np.sqrt(metrics.mean_squared_error(dataTest[:,1],predictedy))
    print("Root mean squares error for lambda=",lambda_,"is",rmse)
```

```python
for i in [2,5,20,50,100]:
    lf=np.zeros((dataTrain.shape[0],i))
    lft=np.zeros((dataTest.shape[0],i))
    for j in range(0,i):
        r=np.random.randint(0,dataTrain.shape[0])
        for k in range(0,dataTrain.shape[0]):
            lf[k,j]=np.exp(-0.1*((np.abs(dataTrain[k][0]-dataTrain[r][0]))**2))
        for k in range(0,dataTest.shape[0]):
            lft[k,j]=np.exp(-0.1*((np.abs(dataTest[k][0]-dataTrain[r][0]))**2))
    w=np.transpose(lf)
    w=np.dot(w,lf)
    w=w+np.multiply(0.1,np.identity(i))
    w=np.linalg.inv(w)
    w=np.dot(w,np.transpose(lf))
    w=w.dot(dataTrain[:,1])
    
    for j in range(0,dataTest.shape[0]):
        predictedy[j]=np.dot(w,lft[j,:])
    fig = plt.figure(figsize=(8,8))
    actual=matplotlib.pyplot.scatter(dataTest[:,0], dataTest[:,1], c='blue')
    predicted=matplotlib.pyplot.scatter(dataTest[:,0], predictedy, c='red')
    plt.legend((actual,predicted),('True output','Predicted output'),loc='upper left')
    plt.title('Plot of training and test points')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    rmse=np.sqrt(metrics.mean_squared_error(dataTest[:,1],predictedy))
    print("Root mean squares error for ",i,"landmark points is",rmse)

    
    
        
```

```python
kmdata=np.loadtxt(fname='kmeans_data.txt',dtype=float,delimiter=None)
fig = plt.figure(figsize=(8,8))
matplotlib.pyplot.scatter(kmdata[:,0], kmdata[:,1], c='blue')
plt.title('Plot of original data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
skmdata=np.square(kmdata)
skmdata=np.divide(skmdata,2.5)
fig = plt.figure(figsize=(8,8))
plt.title('Plot of data having handcrafted features')
plt.xlabel('x1')
plt.ylabel('x2')
matplotlib.pyplot.scatter(skmdata[:,0], skmdata[:,1], c='blue')
plt.show()

#kmeans
c1=skmdata[0,:]
c2=skmdata[1,:]
skmdatadf=pd.DataFrame(skmdata)
z=np.zeros(skmdata.shape[0])
skmdatadf[2]=z
d1=skmdata[0,:]
d2=skmdata[1,:]
c1=[0,0]
c2=[0,0]
#mn=kmdatadf.groupby([2]).mean()
#c=list(mn.loc[0,:])
#print(mn)
#print(c)

while not(np.array_equal(c1,d1)) or not(np.array_equal(c2,d2)):
    c1=d1
    c2=d2
    for i in range(0,skmdata.shape[0]):
        #print(kmdata[i,:],c1)
        if distance.euclidean(skmdata[i,:],c1)<distance.euclidean(skmdata[i,:],c2):
            skmdatadf.loc[i,2]=1
        else:
            skmdatadf.loc[i,2]=2
    #compute new centres
    #print(kmdatadf)
    mn=skmdatadf.groupby([2]).mean()
    #print(mn)
    d1=list(mn.loc[1,:])
    d2=list(mn.loc[2,:])
fig = plt.figure(figsize=(8,8))
colors = ['red','green']
matplotlib.pyplot.scatter(skmdatadf[0], skmdatadf[1], c=skmdatadf[2], cmap=matplotlib.colors.ListedColormap(colors))
plt.title('Clustering of points afer running the kmeans algorithm')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
ld=pd.DataFrame(np.zeros(kmdata.shape[0]))
ld[1]=np.zeros(kmdata.shape[0])
for i in range(0,10):
    r=np.random.randint(0,kmdata.shape[0])
    for j in range(0,kmdata.shape[0]):
        ld.loc[j][0]=np.exp(-0.1*((distance.euclidean(kmdata[r,:],kmdata[j,:]))**2))
    #k meeans
    d1=ld.loc[0,0]
    d2=ld.loc[1,0]
    c1=0
    c2=0
    while not(np.array_equal(c1,d1)) or not(np.array_equal(c2,d2)):
        c1=d1
        c2=d2
        for i in range(0,ld.shape[0]):
            if distance.euclidean(ld.loc[i,0],c1)<distance.euclidean(ld.loc[i,0],c2):
                ld.loc[i,1]=1
            else:
                ld.loc[i,1]=2
        mn=ld.groupby([1]).mean()
    #print(mn)
        d1=list(mn.loc[1,:])
        d2=list(mn.loc[2,:])
    #print(ld[1])
    fig = plt.figure(figsize=(8,8))
    colors = ['red','green']
    matplotlib.pyplot.scatter(kmdata[:,0], kmdata[:,1], c=ld[1], cmap=matplotlib.colors.ListedColormap(colors))
    matplotlib.pyplot.plot(kmdata[r][0],kmdata[r][1],'bo')
    plt.title('Clustering based on landmark based features')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.show()
    
```
