```python
from __future__ import division, print_function, unicode_literals
from sklearn import metrics
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from scipy.spatial import distance


dataTrain=np.loadtxt(fname='ridgetrain.txt',dtype=float,delimiter=None)
dataTest=np.loadtxt(fname='ridgetest.txt',dtype=float,delimiter=None)
K=np.empty((dataTrain.shape[0],dataTrain.shape[0]))
for i in range(0,dataTrain.shape[0]):
    for j in range(0,dataTrain.shape[0]):
        K[i][j]=np.exp(-0.1*((np.abs(dataTrain[i][0]-dataTrain[j][0]))**2))
I=np.identity(dataTrain.shape[0])
for lambda_ in [0.1,1,10,100]:
    alpha=np.multiply(lambda_,I)
    alpha=np.add(alpha,K)
    alpha=np.linalg.inv(alpha)
    alpha=np.dot(alpha,dataTrain[:,1])
    predictedy=np.zeros(dataTest.shape[0])
    for i in range(0,dataTest.shape[0]):
        for j in range(0,dataTrain.shape[0]):
            k=np.exp(-0.1*((np.abs(dataTrain[j][0]-dataTest[i][0]))**2))
            predictedy[i]=predictedy[i]+np.multiply(alpha[j],k)
    fig = plt.figure(figsize=(8,8))
    actual=matplotlib.pyplot.scatter(dataTest[:,0], dataTest[:,1], c='blue')
    predicted=matplotlib.pyplot.scatter(dataTest[:,0], predictedy, c='red')
    plt.legend((actual,predicted),('True output','Predicted output'),loc='upper left')
    plt.title('Plot of training and test points')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    rmse=np.sqrt(metrics.mean_squared_error(dataTest[:,1],predictedy))
    print("Root mean squares error for lambda=",lambda_,"is",rmse)
```
