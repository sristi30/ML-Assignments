```python
from __future__ import division, print_function, unicode_literals
from sklearn import metrics
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from scipy.spatial import distance
import scipy.io
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
#loading data
X=scipy.io.loadmat('mnist_small.mat')
X=X['X']
#PCA
pca = PCA(n_components=2)
Xnew=pca.fit_transform(X)
#kmeans
c=np.zeros((10,2))
print("K means for PCA")
for j in range(10):
    i=0
    for r in np.random.randint(0,10000,10):
        c[i,:]=Xnew[r,:]
        i=i+1
    kmeans = KMeans(n_clusters=10,init=c,algorithm='full').fit(Xnew)
    y=kmeans.predict(Xnew)
    fig = plt.figure(figsize=(5,5))
    matplotlib.pyplot.scatter(Xnew[:,0], Xnew[:,1], c=y, cmap='tab10')
    plt.title('Clustering of points after running the kmeans algorithm on PCA reduced data')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.show()
#tSNE
Xnew=TSNE(n_components=2).fit_transform(X)
print("K means for t-SNE")
for j in range(10):
    i=0
    for r in np.random.randint(0,10000,10):
        c[i,:]=Xnew[r,:]
        i=i+1
    kmeans = KMeans(n_clusters=10,init=c,algorithm='full').fit(Xnew)
    y=kmeans.predict(Xnew)
    fig = plt.figure(figsize=(5,5))
    matplotlib.pyplot.scatter(Xnew[:,0], Xnew[:,1], c=y, cmap='tab10')
    plt.title('Clustering of points after running the kmeans algorithm on tSNE redued data')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.show()
```
