```python
from __future__ import division, print_function, unicode_literals
from sklearn.svm import LinearSVC
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_table('binclass.txt',delimiter=',',header=None)
meanx=data.groupby([2])[0,1].mean()
meanp=meanx.loc[1]
meann=meanx.loc[-1]
datap=data[data[2] ==1]
datan=data[data[2] ==-1]
xp=datap[[0,1]]
xn=datan[[0,1]]
xp=xp.subtract(meanp)
xp=xp.multiply(xp)
xpsum=xp.sum()
xpsum1=xpsum
xpsum=xpsum.divide(xp.shape[0])
sigp=xpsum.sum()
sigp=sigp/xp.shape[1]
xn=xn.subtract(meann)
xn=xn.multiply(xn)
xnsum=xn.sum()
xnsum1=xnsum
xnsum=xnsum.divide(xn.shape[0])
sign=xnsum.sum()
sign=sign/xn.shape[1]
sig0=(xpsum1[0]+xnsum1[0])/data.shape[0]
sig1=(xpsum1[1]+xnsum1[1])/data.shape[0]
sig=(sig0+sig1)/(xp.shape[1])
fig = plt.figure(figsize=(8,8))
colors = ['blue','red']
matplotlib.pyplot.scatter(data[0], data[1], c=data[2], cmap=matplotlib.colors.ListedColormap(colors))
plt.title('Classification of the data of binclass with different covariance matrix of the gaussian class conditional distributions')
plt.xlabel('x1')
plt.ylabel('x2')


x = np.linspace(-12., 30.)   # setting range of x points
y = np.linspace(-1., 35.)[:,None]
plt.contour(x, y.ravel(),0.0677*(x**2)+0.0677*(y**2)-3.5274*x-0.5736*y+33.2451, [0])
plt.show()

fig = plt.figure(figsize=(8,8))
colors = ['blue','red']
matplotlib.pyplot.scatter(data[0], data[1], c=data[2], cmap=matplotlib.colors.ListedColormap(colors))
plt.title('Classification of the data of binclass with same covariance matrix of the gaussian class conditional distributions')
plt.xlabel('x1')
plt.ylabel('x2')


x = np.linspace(-12., 30.)   # setting range of x points
y = np.linspace(-1., 35.)[:,None]
plt.contour(x, y.ravel(),-1.1434*x+1.087*y+1.4699, [0])
plt.show()
```

# 

```python
clf=LinearSVC(dual=False)
clf.fit(data[[0,1]],data[2])

fig = plt.figure(figsize=(8,8))
colors = ['blue','red']
matplotlib.pyplot.scatter(data[0], data[1], c=data[2], cmap=matplotlib.colors.ListedColormap(colors))
plt.title('Classification of the data of binclass using SVM')
plt.xlabel('x1')
plt.ylabel('x2')


x = np.linspace(-12., 30.)   # range of your x points
y = np.linspace(-1., 35.)[:,None]
w=clf.coef_[0]
i=clf.intercept_
plt.contour(x, y.ravel(),w[0]*x+w[1]*y+i, [0])
plt.show()
```
